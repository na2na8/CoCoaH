```python
en_train = pd.read_csv('/home/nykim/2024_spring/00_data/processed_en/en_IMSyPP_train.csv')
en_train = en_train[(en_train['hate']=='0. appropriate')|(en_train['hate']=='1. inappropriate')|
                              (en_train['hate']=='2. offensive')|(en_train['hate']=='3. violent')]
en_valid = pd.read_csv('/home/nykim/2024_spring/00_data/processed_en/en_IMSyPP_valid.csv')
en_valid = en_valid[(en_valid['hate']=='0. appropriate')|(en_valid['hate']=='1. inappropriate')|
                              (en_valid['hate']=='2. offensive')|(en_valid['hate']=='3. violent')]
en_test = pd.read_csv('/home/nykim/2024_spring/00_data/processed_en/en_IMSyPP_test.csv')
en_test = en_test[(en_test['hate']=='0. appropriate')|(en_test['hate']=='1. inappropriate')|
                              (en_test['hate']=='2. offensive')|(en_test['hate']=='3. violent')]
en_test.dropna(subset=['title', 'comment', 'hate'], inplace=True)

it_train = pd.read_csv('/home/nykim/2024_spring/00_data/processed_it/it_IMSyPP_train.csv')
it_train = it_train[(it_train['hate']=='0. appropriato')|(it_train['hate']=='1. inappropriato')|
                              (it_train['hate']=='2. offensivo')|(it_train['hate']=='3. violento')]
it_valid = pd.read_csv('/home/nykim/2024_spring/00_data/processed_it/it_IMSyPP_valid.csv')
it_valid = it_valid[(it_valid['hate']=='0. appropriato')|(it_valid['hate']=='1. inappropriato')|
                              (it_valid['hate']=='2. offensivo')|(it_valid['hate']=='3. violento')]
it_test = pd.read_csv('/home/nykim/2024_spring/00_data/processed_it/it_IMSyPP_test.csv')
it_test = it_test[(it_test['hate']=='0. appropriato')|(it_test['hate']=='1. inappropriato')|
                              (it_test['hate']=='2. offensivo')|(it_test['hate']=='3. violento')]
it_test.dropna(subset=['title', 'comment', 'hate'], inplace=True)
```

```python
print(len(set(en_train['title'])), len(set(en_valid['title'])), len(set(en_test['title'])))
print(len(set(it_train['title'])), len(set(it_valid['title'])), len(set(it_test['title'])))
```
